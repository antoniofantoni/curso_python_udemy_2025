Aqui começaremos cenários reais, com importação de um banco de dados (csv)
Um DataFrame é uma estrutura de dados bidimensional, muito usada na biblioteca pandas do Python.
Ele funciona como uma tabela:
Linhas: representam observações ou registros.
Colunas: representam variáveis ou atributos.

```python
df_pedidos = pd.read_csv('data/pedidos.csv', sep=";")
df_pedidos
```

```python
df_pedidos['Total'] = df_pedidos['Unidade'] * df_pedidos['PrecoUnidade']
```

```python
df_pedidos
```

```python
df_pedidos.head(10)  # Exibe as primeiras 10 linhas do DataFrame
```

```python
vendas_por_regiao = df_pedidos.groupby('Regiao')['Total'].sum().reset_index()# .reset_index() para transformar o resultado em um DataFrame
vendas_por_regiao
```

```python
vendas_por_regiao.plot(x='Regiao', y='Total', kind='bar', legend=True, title='Vendas por Região')
```
