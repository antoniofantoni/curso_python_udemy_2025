```python
notas = {
  'Aluno': ['Antonio', 'Jorge', 'Fabio', 'Alex', 'Joana'],
  'Matemática': [8.5, 7.0, 9.0, 6.5, 8.0],
  'Português': [7.5, 8.0, 6.0, 9.0, 7.0],
  'História': [6.0, 7.5, 8.0, 7.0, 8.5],
  'Geografia': [8.0, 6.5, 7.5, 8.5, 7.0],
  'Ciências': [9.0, 8.5, 7.0, 6.0, 8.0]
}
df_notas = pd.DataFrame(notas)
```

```python
df_notas['Média'] = df_notas[['Matemática', 'Português', 'História', 'Geografia', 'Ciências']].mean(axis=1)
df_notas
```

```python
#alunos aprovados
aprovados = df_notas[df_notas['Média'] >= 7.5]  
aprovados
```

```python
# alunos reprovados
reprovados = df_notas[df_notas['Média'] < 7.6]
reprovados
```
