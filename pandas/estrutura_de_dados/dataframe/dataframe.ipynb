```python
dados_vendas = { # type: ignore
'produto': ['calça', 'camisa', 'sapato', 'tênis', 'jaqueta', 'camisa', 'tênis'],
'preço': [100, 50, 80, 120, 200, 50, 120],
'quantidade': [10, 20, 15, 5, 8, 12, 7],
'mês': ['Janeiro', 'Janeiro', 'Janeiro', 'Janeiro', 'Janeiro', 'Fevereiro', 'Março']
}
df_dados_vendas = pd.DataFrame(dados_vendas)
df_dados_vendas
```

```python
df_dados_vendas['receita'] = df_dados_vendas['preço'] * df_dados_vendas['quantidade']
df_dados_vendas
```

```python
# sumarizando(agrupando) as vendas por produto
resumo_vendas = df_dados_vendas.groupby('produto')['receita'].sum().reset_index()
resumo_vendas
```

```python
#filtrando produtos com receita maior que 1000
produtos_acima_1000 = resumo_vendas[resumo_vendas['receita'] > 1500]
produtos_acima_1000
```

```python
# Plotando os dados
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.bar(resumo_vendas['produto'], resumo_vendas['receita'], color='skyblue')
plt.title('Receita por Produto')
plt.xlabel('Produto')
plt.ylabel('Receita')
plt.xticks(rotation=45)
plt.tight_layout()

# crie codigo abaixo para aparecer rotulo
```
