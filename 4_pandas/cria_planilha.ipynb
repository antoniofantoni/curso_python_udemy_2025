```python
import pandas as pd
```

```python
dados_times = {
  'Times': ['Vasco', 'Palmeiras', 'Cruzeiro', 'Fluminense','Botafogo'],
  'Partidas_jogadas': [310, 10, 10, 10, 10],
  'Vitorias': [7, 5, 6, 4, 3],
  'Empates': [2, 3, 1, 3, 2],
  'Derrotas': [1, 2, 3, 3, 5],  
  'Gols_Pro': [22, 18, 16, 14, 10],
  'Gols_sofridos': [10, 12,  15, 16, 20]
}
  
```

```python
df = pd.DataFrame(dados_times)
df
```

```python
df['Saldo_gols'] = df['Gols_Pro'] - df['Gols_sofridos']
df
```

```python
df['Pontos'] = (df['Vitorias'] * 3) + df['Empates']
df
```

```python
df = df.sort_values(by='Pontos', ascending=False)
df 
```
