```python
df_pedidos = pd.read_csv('Pedidos.csv', sep=';')

```

```python
df_pedidos['Total'] = df_pedidos['Unidade'] * df_pedidos['PrecoUnidade']
```

```python
df_pedidos
```

```python
df_pedidos.head()  # Exibe as 5 primeiras linhas do DataFrame
```

```python
df_pedidos.head(10)  # Exibe as primeiras 10 linhas do DataFrame
```

```python
df_pedidos.tail()  # Exibe as últimas linhas do DataFrame
```

```python
vendas_por_regiao = df_pedidos.groupby('Regiao')['Total'].sum().reset_index()# .reset_index() para transformar o resultado em um DataFrame
vendas_por_regiao.columns = ['Região', 'Total']
vendas_por_regiao
```

```python
import matplotlib.pyplot as plt
vendas_por_regiao.plot(x='Região', y='Total', kind='bar', legend=True, title='Vendas Regionais')
plt.xticks(rotation=45)
```

```python
# Operações
import pandas as pd
import matplotlib.pyplot as plt

data = {
  'Nome': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
  'Idade': [25, 30, 35, 40, 45],
  'Cargo': ['Analista', 'Gerente', 'Diretor', 'Analista', 'Assistente'],
  'Salario': [5000, 7000, 10000, 6000, 4000]
}
```

```python
df = pd.DataFrame(data)
df
df.info()  # Exibe informações sobre o DataFrame, como tipos de dados e contagem de valores não nulos
df.describe()  # Exibe estatísticas descritivas para colunas numéricas
```

```python
df['Salario']  # Filtra só os salarios
```

```python
df[['Salario']].rename(columns={'Salario': 'Salários'})
```

```python
#Ordenem os dados por idade
df.sort_values(by='Idade', ascending=True)  # Ordena por idade em ordem crescente 
```

```python
df['Bonus'] = df['Salario'] * 0.1  # Cria uma nova coluna 'Bonus' com 10% do salário
df
```

```python
df
```

```python
df.groupby('Cargo')['Salario'].mean().reset_index()  # Agrupa por cargo e calcula a média salarial
```

```python
df.plot(
  x='Nome', 
  y='Salario', 
  kind='bar',
  title='Salário por Funcionário',
  rot=45
)
```
